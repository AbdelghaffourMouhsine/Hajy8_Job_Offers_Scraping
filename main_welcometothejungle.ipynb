```python
def extract_jobs(self):
    job_items = self.get_element('//ul[@data-testid="search-results"]/li[@data-testid="search-results-list-item-wrapper"]', group=True)
    if job_items["status"]:
        job_items = job_items["data"]
        # print(len(job_items))
        jobs = []
        for job_item in job_items:
            jobs.append(extract_job(self, job_item)) ####################################
            # display(jobs[-1])
        return jobs
    else:
        print({"status": False, "data":job_items["data"] })
    time.sleep(2)

# job_items = extract_jobs(self)
```

```python
def extract_job(self, from_elem):
    job = {}
    job_link = self.get_element('div/div/a', from_elem=from_elem)
    if job_link["status"]:
        job_link = job_link["data"]
        job['job_url'] = job_link.get_attribute('href')

    job_title = self.get_element('div/div/div/div[2]/a/h4', from_elem=from_elem)
    if job_title["status"]:
        job_title = job_title["data"]
        job['job_title'] = job_title.get_attribute('innerText')

    job_location = self.get_element('div/div/div/div[2]/div/div/p/span/span', from_elem=from_elem)
    if job_location["status"]:
        job_location = job_location["data"]
        job['job_location'] = job_location.get_attribute('innerText')

    job_more_inf = self.get_element('div/div/div/div[3]', from_elem=from_elem)
    if job_more_inf["status"]:
        job_more_inf = job_more_inf["data"]
        job['job_inf'] = job_more_inf.get_attribute('innerText')

    job_date = self.get_element('div/div/div/div[4]', from_elem=from_elem)
    if job_date["status"]:
        job_date = job_date["data"]
        job['job_date'] = job_date.get_attribute('innerText')

    time.sleep(2)
    return job
    
# jobs = []
# for job_item in job_items:
#     jobs.append(extract_job(self, job_item))
#     display(jobs[-1])
```

```python
def extract_company_info(self, job_url):
    company = {}
    self.driver.get(job_url)
    time.sleep(3)
    company_linkedin = self.get_element('//*[@id="app"]/div/div/div/div/div[3]/aside/div[2]/div/a[contains(@href, "linkedin")]')
    if company_linkedin["status"]:
        company_linkedin = company_linkedin["data"]
        company['company_linkedin'] = company_linkedin.get_attribute('href')
    else:
        print(company_linkedin)

    company_web_site = self.get_element('//*[@id="the-company-section"]/div/div[2]/ul/li[1]/a')
    if company_web_site["status"]:
        company_web_site = company_web_site["data"]
        company['company_website'] = company_web_site.get_attribute('href')
    else:
        print(company_web_site)

    company_name = self.get_element('//*[@id="app"]/div/div/div/div/div[3]/section/div[1]/div[1]/a/div/span')
    if company_name["status"]:
        company_name = company_name["data"]
        company['company_name'] = company_name.get_attribute('innerText')
    else:
        print(company_name)
        
    return {'status' : True, 'data': company}

# for job in jobs[20:]:
#     print(job | extract_company_info(self, job['job_link']))
```

```python
def get_next_page(self):
    first_url = self.driver.current_url
    current_url = self.driver.current_url
    i=0
    while first_url == current_url and i<10:
        try:
            next_button = self.get_element('//nav[@aria-label="Pagination"]/ul/li/a', group=True)
            if next_button["status"]:
                next_button = next_button["data"][-1]
                disabled = next_button.get_attribute('aria-disabled')
                if disabled == 'false':
                    self.click_elem(next_button)
                else:
                    return 'fin'
            else:
                print('error')
        except Exception as e:
            # print('Error at get_next_page : ',e)
            pass
        current_url = self.driver.current_url
        i = i+1
    if first_url != current_url:
        return True
    else:
        return False
        
# get_next_page(self)
```
